```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import pickle
import matplotlib.pyplot as pl
import time
from tqdm import tqdm
```

```python
ssc_map = pd.read_csv('ssc/SSC_2016_AUST.csv')
keep = ['SSC_CODE_2016', 'SSC_NAME_2016', 'STATE_CODE_2016',
        'STATE_NAME_2016']
ssc_map = ssc_map[keep]
ssc_map = ssc_map[ssc_map['STATE_NAME_2016'] == 'New South Wales']
ssc_map = ssc_map.drop_duplicates()
ssc_map.head()
```

```python
url_format = 'https://quickstats.censusdata.abs.gov.au/census_services/getproduct/census/{}/quickstat/SSC{}'
html = {}
if 'SSC_NAME_2016' in ssc_map:
    ssc_map = ssc_map.set_index('SSC_NAME_2016')
ssc = ssc_map['SSC_CODE_2016'].to_dict()
# ssc
```

```python
url_format = 'https://quickstats.censusdata.abs.gov.au/census_services/getproduct/census/{}/quickstat/SSC{}'
year='2016'
for suburb, ssc_num in tqdm(ssc.items()):
    if suburb in html or suburb == 'Mogilla':
        continue
    print(suburb)
    url = url_format.format(year, ssc_num)
    resp = requests.get(url)
    txt = resp.text
    html[suburb] = txt
    time.sleep(2)

with open('allsource2016.pkl', 'wb') as f:
    source = pickle.dump(html, f)
```
